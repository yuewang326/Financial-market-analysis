```python
# query daily measure table and determine max closing price w/ date
connection = sql.connect('dj.db')
cursor = connection.cursor()
query = '''
SELECT date, closing
FROM daily_measures
WHERE ticker_id=3
ORDER BY date
'''
cursor.execute(query)

maxPrice = 0
maxDate = None
for row in cursor: # iterate through rows using the cursor
    date = row[0]
    closing = row[1]
    #print (date, closing)
    if closing > maxPrice:
        maxPrice = closing
        maxDate = date
print(maxDate, maxPrice)
```

```python
# query daily measure table to determine max closing price % change
connection = sql.connect('dj.db')
cursor = connection.cursor()
query = '''
SELECT date, closing
FROM daily_measures
WHERE ticker_id=3
ORDER BY date
'''
cursor.execute(query)
rows = cursor.fetchall()
df = pd.DataFrame(rows, columns = ['date', 'closing'])
df.set_index("date", inplace=True, drop=True)
df
```

```python
#  cont: add percent change column
df["pctchg"] = df.closing.pct_change()
df
```

```python
#  (cont): query by using max() function to find largest 1-day % gain row
df[df.pctchg == df.pctchg.max()]
```

```python
# (cont): query by using min() function to find larget 1-day % loss
df[df.pctchg == df.pctchg.min()]
```
